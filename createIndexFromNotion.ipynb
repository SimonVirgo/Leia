```python
import logging
import sys

logging.basicConfig(stream=sys.stdout, level=logging.INFO)
logging.getLogger().addHandler(logging.StreamHandler(stream=sys.stdout))
```

```python

# Open and read the config file
with open('config.json', 'r') as config_file:
    config_data = json.load(config_file)

# Retrieve the API key from the config data
api_key = config_data['api_key']
os.environ['OPENAI_API_KEY'] = api_key

# Retrieve the Notion integration token from the config data
notion_integration_token = config_data['notion_key']
```

```python
from llama_index import ListIndex, NotionPageReader
database_id = "1ec3c41e9e914daeae0741c11cf81364"

# https://developers.notion.com/docs/working-with-databases for how to find your database id

documents = NotionPageReader(integration_token=notion_integration_token).load_data(
    database_id=database_id
)

print(documents)
```
